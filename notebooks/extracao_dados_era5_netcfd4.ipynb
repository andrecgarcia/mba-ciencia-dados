```python
import xarray as xr
import pandas as pd

# Caminho para o arquivo no Google Drive
caminho_arquivo = "/content/drive/MyDrive/data_stream-oper_stepType-accum.nc"

# Abre o arquivo NetCDF
ds = xr.open_dataset(caminho_arquivo)

# Coordenadas da estação (latitude e longitude)
lat_target, lon_target = -22.022777777778, -43.960555555556

# Seleciona o ponto mais próximo disponível na grade
ponto = ds.sel(latitude=lat_target, longitude=lon_target, method="nearest")

# Extrai a variável de precipitação total
serie_tp = ponto['tp'].to_dataframe().reset_index()

# Converte de metros para milímetros
serie_tp['tp_mm'] = serie_tp['tp'] * 1000

# Mostra as 10 primeiras linhas
print(serie_tp.head(10))

# (Opcional) Exporta para CSV
serie_tp.to_csv("/content/drive/MyDrive/serie_tp_crhea_2007.csv", index=False)
```

```python
# Caminho para o arquivo no Google Drive
caminho_arquivo = "/content/drive/MyDrive/data_stream-oper_stepType-instant.nc"

# Abre o arquivo NetCDF
ds = xr.open_dataset(caminho_arquivo)

# Coordenadas da estação (latitude e longitude)
lat_target, lon_target = -22.022777777778, -43.960555555556

# Seleciona o ponto mais próximo disponível na grade
ponto = ds.sel(latitude=lat_target, longitude=lon_target, method="nearest")

# Extrai a série temporal da temperatura (t2m)
serie_temp = ponto['t2m'].to_dataframe().reset_index()

# Converte de Kelvin para Celsius
serie_temp['t2m_celsius'] = serie_temp['t2m'] - 273.15

# Mostra as 10 primeiras linhas
print(serie_temp.head(10))

serie_temp.to_csv("/content/drive/MyDrive/serie_temp_crhea_2007.csv", index=False)
```
